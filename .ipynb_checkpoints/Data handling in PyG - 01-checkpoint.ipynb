```python
import numpy as np
import torch
import torch_geometric.datasets as datasets
import torch_geometric.data as data
import torch_geometric.loader as loader
import torch_geometric.transforms as transforms
import networkx as nx
from torch_geometric.utils.convert import to_networkx
import matplotlib.pyplot as plt
```

# Data 

Creating a dummy graph

```python
node_features = torch.rand((100, 16), dtype=torch.float) # #nodes = 100, dimension = 16
```

```python
rows = np.random.choice(100, 500)
cols = np.random.choice(100, 500) # #edges = 500
edges = torch.tensor([rows, cols])
print(edges.shape) # stored in COO format
```

```python
edges_attr = np.random.choice(3, 500)
print(edges_attr.shape)
# print(edges_attr)
```

```python
t = torch.rand((100)).round().long()
```

## Converting the graph information into a PyG Data object

```python
graph = data.Data(x=node_features, edge_index=edges, edge_attr=edges_attr, y=t)
```

```python
graph
```

## Visualize the information in the graph (data object)

```python
for prop in graph:
    print(prop)
```

```python
vis = to_networkx(graph)

node_labels = graph.y.numpy()

plt.figure(1, figsize=(15, 10))
nx.draw(vis, cmap=plt.get_cmap('Set3'), node_color=node_labels, node_size=70, linewidths=6)
plt.show()
```

## Batch

With the Batch object we can represent multiple graphs as a single disconnected graph

```python
graph_2 = graph
```

```python
batch = data.Batch().from_data_list([graph, graph_2])
```

```python
print("Number of graphs: ", batch.num_graphs)
print("Graph at index 1: ", batch[1])
print("Retrieve the list of graphs:\n", len(batch.to_data_list()))
```

## Cluster

ClusterData groups the nodes of a graph into a specific number of cluster for faster computation in large graphs, the use ClusterLoader to load batches of clusters.

```python
cluster = loader.ClusterData(graph, 5)
```

```python
clusterloader = loader.ClusterLoader(cluster)
```

## Sampler

For each convolutional layer, sample a maximum of nodes from each neighbourhood (like GraphSAGE)

```python
sampler = loader.NeighborSampler(graph.edge_index, sizes=[3, 10], batch_size=4 , shuffle=False)
```

```python
for s in sampler:
    print(s)
    print("===============\n")
    break
```

```python
print("Batch size: ", s[0])
print("Number of unique node involved in the sampling: ", len(s[1]))
print("Number of neighbours sampled: ", len(s[2][0].edge_index[0]), len(s[2][1].edge_index[0]))
```

```python
name = 'Cora'
transform = transforms.Compose([
    transforms.RandomNodeSplit('train_rest', num_val=500, num_test=500),
    transforms.TargetIndegree(),
])
cora = datasets.Planetoid('./data', name, pre_transform=transforms.NormalizeFeatures(), transform=transform)
```

```python
aids = datasets.TUDataset(root="./data", name="AIDS")
```

```python
print("Cora info:")
print('# of graphs:', len(cora))
print('# Classes (nodes)', cora.num_classes)
print('# Edge features', cora.num_edge_features)
print('# Node features', cora.num_node_features)
```
